Percolation on $\mathbb{Z}^2$ 

The goal of the second task is to simmulate an edge percolation on $\mathbb{Z}^2$.
 * Simmulate an edge percolation in a $n \times n$ box with parameter $p$.
 * Plot a visualization of the open/closed edges.
 * Identify connected components.
 * Draw each connected component with a different color.

```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
import random

# seed the pseudorandom number generator
from random import seed
from random import random
# seed random number generator
seed(2024)

# Parameters
n = 100  # Size of the grid
p = 0.50  # Percolation probability

# Create a grid graph
G = nx.grid_2d_graph(n, n)

# Randomly remove edges with probability 1-p
for edge in list(G.edges):
    if np.random.rand() > p:
        G.remove_edge(*edge)

# Find connected components
components = list(nx.connected_components(G))

# Assign a color to each component
colors = plt.cm.nipy_spectral(np.linspace(0, 1, len(components)))

# Draw the graph
pos = {(x, y): (x, y) for x, y in G.nodes()}
plt.figure(figsize=(12, 12))

for i, component in enumerate(components):
    nx.draw_networkx_edges(
        G.subgraph(component), pos=pos, edge_color=[colors[i]], width=2
    )

nx.draw_networkx_nodes(G, pos=pos, node_size=0.1, node_color="black")

plt.title(f"Edge Percolation on Z2 Grid (n={n}, p={p})")
plt.axis("equal")
plt.show()
```
